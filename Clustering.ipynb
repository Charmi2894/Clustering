# Theory Questions

1. What is unsupervised learning in the context of machine learning?

- Unsupervised learning is a category of machine learning where the algorithm is trained on data that has not been labeled or classified. The goal is to explore the data's inherent structure, discover hidden patterns, and draw inferences without a predefined target variable or explicit guidance.
- Clustering is a classic example of unsupervised learning.

2. How does K-Means clustering algorithm work?
- K-Means clustering is an iterative algorithm that partitions N data points into K predefined, non-overlapping clusters
  1. Initialization: Select K initial centroids.
  2. Assignment Step: Assign every data point to the cluster whose centroid is the closest.
  3. Update Step: Recalculate the centroids by taking the mean of all data points assigned to that cluster.
  4. Convergence: Repeat steps 2 and 3 until the centroids no longer move significantly or a maximum number of iterations is reached.

3. Explain the concept of a dendrogram in hierarchical clustering.
- A dendrogram is a tree-like diagram used to illustrate the arrangement of clusters produced by hierarchical clustering.
  - Vertical Lines represent the clusters being merged or split.
  - Horizontal Lines indicate the distance or dissimilarity between the merged clusters (the linkage distance).
  - The point at which two branches join represents the formation of a cluster, and the height of the union indicates the distance between those two clusters.
  - To determine the final clusters, you cut the dendrogram at a specific height.

4. What is the main difference between K-Means and Hierarchical Clustering?

- K-Means: Is a partitioning method that creates non-overlapping clusters based on a pre-specified number, K. It produces spherical-shaped clusters.
- Hierarchical Clustering: Is a connectivity-based method that builds a hierarchy of clusters. It can be agglomerative or divisive. It does not require specifying the number of clusters beforehand, which can be determined later by cutting the dendrogram.

5. What are the advantages of DBSCAN over K-Means?

- Non-spherical Clusters: It can find clusters of arbitrary shapes, unlike K-Means which is restricted to convex or spherical shapes.
- Outlier Detection: It naturally identifies noise points that do not belong to any cluster.
- No Pre-defined K: It does not require the user to specify the number of clusters, K, in advance.

6. When would you use Silhouette Score in clustering?

- You would use it:To compare and choose the optimal number of clusters (K) for algorithms like K-Means or Agglomerative Clustering.
- To assess how well each data point fits within its assigned cluster versus how far it is from other clusters.
- As a better evaluation metric than just inertia because it accounts for both intra-cluster distance and inter-cluster distance.

7. What are the limitations of Hierarchical Clustering?
- Computational Cost: For large datasets, the memory and time complexity (often $O(N^3)$ or $O(N^2)$) can be prohibitive compared to K-Means ($O(N \cdot K \cdot I)$).
- No Redo: Once a cluster is formed or points are merged, the action cannot be undone, which can be sensitive to noise and outliers.
- Difficulty with Large Datasets: It scales poorly and can be hard to interpret for very large numbers of data points.

8. Why is feature scaling important in clustering algorithms like K-Means?

- Feature scaling is crucial because K-Means relies on the Euclidean distance between points and centroids.
- If features are on different scales, the feature with the larger magnitude will dominate the distance calculation.
- Scaling ensures all features contribute equally to the distance calculation, preventing bias and leading to more meaningful, accurate clusters.

9. How does DBSCAN identify noise points?
- DBSCAN identifies a data point as noise if it is neither a core point nor a border point.
- A noise point is a point that has fewer than min_samples within its eps (epsilon) radius.
- It belongs to no cluster and is effectively isolated in a low-density region of the feature space.

10. Define inertia in the context of K-Means.

- Inertia is a metric that measures the compactness of the clusters.
- It is calculated as the sum of the squared distances between each point and the centroid of the cluster to which it belongs.
- A lower inertia value indicates that the clusters are more dense and the points are closer to their respective centroids.

11. What is the elbow method in K-Means clustering?
- The elbow method is an empirical technique used to estimate the optimal number of clusters ($K$) for K-Means.
  - The algorithm is run for a range of $K$ values (e.g., $K=1$ to 10).
  - The inertia (WCSS) is calculated for each run.
  - The inertia values are plotted against $K$.
  - The optimal $K$ is typically chosen at the "elbow" of the plotâ€”the point where the decrease in inertia starts to slow down significantly.


12. Describe the concept of "density" in DBSCAN.

- In DBSCAN, "density" is defined based on two parameters: $\epsilon$ (epsilon) and min_samples.
  - A region of space is considered "dense" if, for a given point, there are at least min_samples data points within a distance of $\epsilon$.
  - A point satisfying this condition is called a core point, and clusters are formed by chaining together these high-density core points.

13. Can hierarchical clustering be used on categorical data?

- Yes, hierarchical clustering can be used on categorical data, but it requires transforming the data and using an appropriate distance/dissimilarity measure.
  - Transformation: Categorical variables must first be converted into a numerical format, often using methods like one-hot encoding.
  - Distance Metric: Instead of Euclidean distance, metrics more suited for categorical data, such as Jaccard distance or Hamming distance, should be used for the linkage calculation.

14. What does a negative Silhouette Score indicate?
- The Silhouette Score ranges from -1 to +1.
- A score close to +1 indicates the point is well-clustered, as it is far from neighboring clusters.
- A score close to 0 indicates the point is on or very close to the decision boundary between two clusters.
- A negative Silhouette Score indicates that the data point has been assigned to the wrong cluster.


15. Explain the term "linkage criteria" in hierarchical clustering.

- Linkage criteria (or linkage type) define the metric used to calculate the distance between two existing clusters (or groups of points) when deciding which clusters to merge in agglomerative clustering.
- Common criteria include:
  - Ward: Minimizes the variance within each of the merged clusters.
  - Single: Uses the minimum distance between any point in the first cluster and any point in the second cluster.
  - Complete: Uses the maximum distance between any point in the first cluster and any point in the second cluster.
  - Average: Uses the average distance between all pairs of points across the two clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

- Varying Size/Density: If clusters are of different sizes or densities, the mean-based centroid may be pulled toward the center of a larger or denser cluster, causing smaller/sparser clusters to be incorrectly grouped or points to be incorrectly assigned.
- Global Boundary: The assignment step creates straight-line (Voronoi) decision boundaries, which cannot separate non-spherical or irregularly sized clusters effectively.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

- The two core parameters in DBSCAN are:
  - $\epsilon$ (epsilon or eps): The maximum radius of the neighborhood around a data point.
      - Influence: A larger $\epsilon$ leads to larger neighborhoods, potentially merging multiple distinct clusters into one, or including more noise points. A smaller $\epsilon$ can split a single cluster into multiple smaller clusters.
  - min_samples: The minimum number of points required within the $\epsilon$ radius to form a dense region.
    - Influence: A larger min_samples makes the algorithm stricter about density, leading to fewer clusters and more points being classified as noise. A smaller min_samples can make noise points appear to be part of a cluster.

18. How does K-Means++ improve upon standard K-Means initialization?

- Standard K-Means is highly susceptible to a poor choice of initial centroids, which can lead to slow convergence or a sub-optimal local minimum. K-Means++ is an intelligent initialization technique that aims to solve this.
  - Improvement: K-Means++ selects the initial centroids such that they are far apart from each other.
  - Process: After the first centroid is chosen randomly, subsequent centroids are selected with a probability proportional to the square of the distance from the point to the nearest existing centroid. This greatly increases the chance of finding a globally optimal clustering solution.

19. What is agglomerative clustering?
- Agglomerative clustering is a type of hierarchical clustering that follows a "bottom-up" approach.
- Process: It starts by treating each data point as a single cluster. It then iteratively merges the two closest clusters based on a specified linkage criteria.
- Result: This merging process continues until all data points are grouped into a single, comprehensive cluster. The entire process is represented by a dendrogram.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
- While both measure cluster quality, Inertia (WCSS) only measures intra-cluster distance. The inertia value inherently decreases as the number of clusters ($K$) increases, making it a poor sole indicator of the optimal $K$.
- Silhouette Score is a better metric because it considers both:
  - Intra-cluster cohesion ($a$): How close a point is to others in its own cluster.
  - Inter-cluster separation ($b$): How far a point is from the points in the next nearest cluster.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs, make_moons, make_circles, load_iris, load_wine, load_breast_cancer, load_digits
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib.colors import ListedColormap

# Set a consistent style and random seed for reproducibility
sns.set_style("whitegrid")
np.random.seed(42)
```

```python
#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.
# 1. Generate synthetic data
X_blobs, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.8, random_state=42)

# 2. Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_blobs)

# 3. Visualize
plt.figure(figsize=(8, 6))
# Create a colormap
cmap = ListedColormap(sns.color_palette("Set1", n_colors=4).as_hex())
scatter = plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_kmeans, s=50, cmap=cmap)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.8, marker='X', label='Centroids')
plt.title('K-Means Clustering with 4 Centers (make_blobs)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.
# 1. Generate synthetic data
X_moons, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

# 2. Apply DBSCAN (Optimal params for this data)
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X_moons)

# 3. Identify noise points (label -1)
noise_mask = (y_dbscan == -1)
clusters_mask = (y_dbscan != -1)

# 4. Visualize
plt.figure(figsize=(8, 6))
# Plot clusters
plt.scatter(X_moons[clusters_mask, 0], X_moons[clusters_mask, 1],
            c=y_dbscan[clusters_mask], cmap='viridis', s=50, label='Cluster Points')
# Highlight outliers
plt.scatter(X_moons[noise_mask, 0], X_moons[noise_mask, 1],
            color='red', marker='x', s=100, label='Outliers (Noise)')

plt.title('DBSCAN Clustering on make_moons Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.
# 1. Load the Wine dataset and standardize features
wine = load_wine()
X_wine = wine.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_wine)

# 2. Apply K-Means (The true number of classes is 3)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_scaled)

# 3. Create a DataFrame for easy counting
df_wine = pd.DataFrame(X_scaled, columns=wine.feature_names)
df_wine['Cluster'] = y_kmeans

# 4. Print the size of each cluster
cluster_sizes = df_wine['Cluster'].value_counts().sort_index()
print("Size of each K-Means cluster:")
print(cluster_sizes)
```

```python
#25. Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result.
# 1. Generate synthetic data
X_circles, y_true = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# 2. Apply DBSCAN (DBSCAN handles circular data well with appropriate params)
dbscan = DBSCAN(eps=0.1, min_samples=5)
y_dbscan = dbscan.fit_predict(X_circles)

# 3. Visualize
plt.figure(figsize=(8, 6))
# Exclude noise points (-1) from the main cluster plot
unique_labels = set(y_dbscan)
core_samples_mask = np.zeros_like(y_dbscan, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

cmap = plt.cm.get_cmap('Spectral')
for k in unique_labels:
    if k == -1:
        # Black color for noise points
        col = 'k'
        marker = 'x'
        label = 'Noise'
    else:
        col = cmap(k / len(unique_labels))
        marker = 'o'
        label = f'Cluster {k}'

    class_member_mask = (y_dbscan == k)
    xy = X_circles[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], marker, markerfacecolor=col,
             markeredgecolor='k', markersize=6, label=label)

plt.title('DBSCAN Clustering on make_circles Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for $k=2$ to 5. Display as a bar chart.
# 1. Generate synthetic data (assuming 3 true centers)
X_blobs, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# 2. Evaluate Silhouette Scores for k=2 to 5
K_range = range(2, 6)
silhouette_scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    y_pred = kmeans.fit_predict(X_blobs)
    score = silhouette_score(X_blobs, y_pred)
    silhouette_scores.append(score)

# 3. Display as a bar chart
plt.figure(figsize=(8, 6))
plt.bar([str(k) for k in K_range], silhouette_scores, color='skyblue')
plt.title('Silhouette Scores for K-Means Clustering ($K=2$ to 5)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.ylim(min(silhouette_scores) - 0.05, 1)
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
# 1. Generate synthetic data with overlapping clusters (high std)
X_overlap, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# 2. Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_overlap)
y_kmeans = kmeans.predict(X_overlap)

# 3. Visualize with decision boundaries

# Create meshgrid
h = 0.02 # step size in the mesh
x_min, x_max = X_overlap[:, 0].min() - 1, X_overlap[:, 0].max() + 1
y_min, y_max = X_overlap[:, 1].min() - 1, X_overlap[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict cluster for each point in the mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
# Plot the decision boundaries
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Pastel1, aspect='auto', origin='lower', alpha=0.8)

# Plot the data points
plt.scatter(X_overlap[:, 0], X_overlap[:, 1], c=y_kmeans, s=50, cmap=plt.cm.Set1, edgecolor='k')

# Plot the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.9, marker='X', label='Centroids')

plt.title('K-Means Clustering on Overlapping Blobs with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#34. Load the Breast Cancer dataset and compare inertia values for $K=2$ to 6 using K-Means. Show results in a line plot.
# 1. Load data and scale features
cancer = load_breast_cancer()
X_cancer = cancer.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cancer)

# 2. Calculate inertia for K=2 to 6
K_range = range(2, 7)
inertia_values = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# 3. Show results in a line plot (Elbow Method)
plt.figure(figsize=(8, 6))
plt.plot(K_range, inertia_values, marker='o', linestyle='-', color='b')
plt.title('K-Means Inertia (WCSS) vs. Number of Clusters (K)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (WCSS)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.
from sklearn.metrics import silhouette_samples

# 1. Load the Iris dataset and apply K-Means (K=3)
iris = load_iris()
X_iris = iris.data
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_iris)

# 2. Calculate the Silhouette Coefficient for each sample
sample_silhouette_values = silhouette_samples(X_iris, y_kmeans)
silhouette_avg = silhouette_score(X_iris, y_kmeans)

# 3. Plot the Silhouette Coefficients
fig, ax1 = plt.subplots(1, 1, figsize=(8, 6))
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, len(X_iris) + (3 + 1) * 10]) # For padding

y_lower = 10
for i in range(3): # For each cluster
    # Aggregate the silhouette scores for samples belonging to cluster i
    ith_cluster_silhouette_values = sample_silhouette_values[y_kmeans == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.Spectral(float(i) / 3)
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with their cluster numbers
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10 # 10 for the 0 samples

ax1.axvline(x=silhouette_avg, color="red", linestyle="--", label=f"Average Score ({silhouette_avg:.2f})")
ax1.set_title("Silhouette Plot for K-Means Clustering on Iris (K=3)")
ax1.set_xlabel("Silhouette Coefficient Values")
ax1.set_ylabel("Cluster label")
ax1.legend()
plt.show()
```
